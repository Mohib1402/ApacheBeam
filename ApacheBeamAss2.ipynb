```python
import pandas as pd
import sweetviz as sv

# Load the dataset
df = pd.read_csv('/content/Telco-Customer-Churn.csv')

# Generate the Sweetviz report
report = sv.analyze(df)

# Save and display the report
report.show_html(filepath='Telco_Churn_Sweetviz_Report.html')

```

```python
# Generate Sweetviz report with the target column 'Churn'
report = sv.analyze(df, target_feat='Churn')
report.show_html(filepath='Telco_Churn_Targeted_Sweetviz_Report.html')
```

```python
# Create subsets for churned and non-churned customers
churned = df[df['Churn'] == 'Yes']
not_churned = df[df['Churn'] == 'No']

# Generate Sweetviz comparison report
report = sv.compare([churned, "Churned Customers"], [not_churned, "Non-Churned Customers"])
report.show_html(filepath='Telco_Churn_Comparison_Report.html')
```
